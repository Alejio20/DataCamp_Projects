```python
# Import libraries here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
# Read csv file
df = pd.read_csv('recipe_site_traffic_2212.csv')
df.head()
```

```python
df.info()
```

```python
# Replace 'Chicken Breast' with 'Chicken' in the category column
df['category'] = df['category'].replace('Chicken Breast', 'Chicken')
```

```python
# Replace '4 as a snack' with '4' in the servings column
df['servings'] = df['servings'].replace('4 as a snack', '4')
# Replace '6 as a snack' with '6' in the servings column
df['servings'] = df['servings'].replace('6 as a snack', '6')
# Convert servings column to integer type
df['servings'] = df['servings'].astype(int)
```

```python
# Fill missing value with 'Low' in the high_traffic column
df['high_traffic'] = df['high_traffic'].fillna('Low')
```

```python
# Drop all row with missing values
df.dropna(inplace=True)
df.info()
```

```python
# Convert the categorical variables into numeric
from sklearn.preprocessing import LabelEncoder

# Create an instance of LabelEncoder
labelencoder = LabelEncoder()

# Fit ad transform variables
df['category'] = labelencoder.fit_transform(df['category'])
df['high_traffic'] = labelencoder.fit_transform(df['high_traffic'])
```

```python
# Separate the features and target variable
X = df.drop(columns=['high_traffic', 'recipe'], axis=1)  # Features (input variables)
y = df['high_traffic']  # Target variable
```

```python
# Normalize the numeric variables
from sklearn.preprocessing import StandardScaler

# Create an instance of StandardScaler
scaler = StandardScaler()

# Fit ad transform variables
X[['calories']] = scaler.fit_transform(X[['calories']])
X[['carbohydrate']] = scaler.fit_transform(X[['carbohydrate']])
X[['sugar']] = scaler.fit_transform(X[['sugar']])
X[['protein']] = scaler.fit_transform(X[['protein']])
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=54)
```

```python
# K-Nearest Neighbor Model
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

param = {'n_neighbors': np.arange(1,500)}

knn = KNeighborsClassifier()
grid_knn = GridSearchCV(estimator=knn, param_grid=param, scoring='accuracy', cv=5, n_jobs=-1)
grid_knn.fit(X_train, y_train)

# Extract the best model
best_k = grid_knn.best_params_['n_neighbors'] # best n_neighbors value
best_knn = grid_knn.best_estimator_ # best knn model

# Evaluate the Accuracy on test set
best_knn_acc = best_knn.score(X_test, y_test)

print('The best KNN model has n_neighbor of ', best_k)
print('The best KNN Accuracy score is ', best_knn_acc)
```

```python
# Decision Tree Classifier Model
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

param_dt = {
    'max_depth': np.arange(0,15),
    'min_samples_split': np.arange(0,15),
    'min_samples_leaf': np.arange(0,15),
    'max_features': ['auto', 'sqrt', 'log2']
}

dt = DecisionTreeClassifier(random_state=42)
grid_dt = GridSearchCV(estimator=dt, param_grid=param_dt, scoring='accuracy', cv=5, n_jobs=-1)
grid_dt.fit(X_train, y_train)

# Extract the best model
best_dt_params = grid_dt.best_params_
best_dt = grid_dt.best_estimator_

# Evaluate the Accuracy on test set
best_dt_acc = best_dt.score(X_test, y_test)

print('The best DecisionTree model has params as follows:', best_dt_params)
print('The best DecisionTree Accuracy score is ', best_dt_acc)
```

```python
# For XGBoost Classifier Model
from sklearn.model_selection import GridSearchCV
import xgboost as xgb

# Define a parameter grid for GridSearchCV
param_xgb = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'n_estimators': [50, 100, 200],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

# Define the XGBoost classifier
xg_clf = xgb.XGBClassifier()

# Instantiate GridSearchCV
grid_search = GridSearchCV(estimator=xg_clf, param_grid=param_xgb, cv=5, scoring='accuracy', verbose=1, n_jobs=-1)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Get the best parameters from the grid search
best_xgb_params = grid_search.best_params_

# Use the best model to make predictions
best_xgb = grid_search.best_estimator_

# Evaluate the Accuracy on test set
best_xgb_acc = best_xgb.score(X_test, y_test)

print('The best XGBoost classifier model has params as follows:', best_xgb_params)
print('The best XGBoost classifier Accuracy score is ', best_xgb_acc)
```
